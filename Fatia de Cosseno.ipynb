```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

# if using a jupyter notebook
%matplotlib inline    

```

```python
x = np.arange(0,2*np.pi,0.1)   # start,stop,step
y = np.sin(x)

figure(figsize=(8,4))

plt.axvline(x=0)
plt.axvline(x=1)
plt.axvline(x=2)
plt.axvline(x=3)
plt.axvline(x=4)
plt.axvline(x=5)
plt.axvline(x=6)

x_ = np.array([i for i in range(7)])
y_ = np.array(np.sin(x_))


plt.plot(x,y)
plt.plot(x_, y_, 'ro')
plt.show()
```
